```python
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
driver=webdriver.Chrome(r"C:\Users\pop\Downloads\chromedriver_win32\chromedriver.exe")
```

```python
driver.get("https://www.amazon.in/gp/bestsellers/books/")
elements = driver.find_elements(By.XPATH,"//div[@class='a-column a-span12 a-text-center _cDEzb_grid-column_2hIsc']")
print (len(elements))
for i in elements :
    print(i.text)
```

```python
titles=[]
prices=[]

for product in elements :
    try:
        title = product.find_element(By.XPATH,".//div[@class='_cDEzb_p13n-sc-css-line-clamp-1_1Fn1y']").text
        price = product.find_element(By.XPATH,".//span[@class='_cDEzb_p13n-sc-price_3mJ9Z']").text
    except:
        print('error')
    titles.append(title)
    prices.append(price)
    
print (len(prices))
```

```python
import csv
with open("exported4.csv", "w" ,encoding ='utf-8', newline= '') as f:
    csvwriter=csv.writer(f, delimiter=',')
    csvwriter.writerow(["title", 'price'])
    for t in range(0,len(titles)):
        csvwriter.writerow([titles[t],prices[t]])
```
